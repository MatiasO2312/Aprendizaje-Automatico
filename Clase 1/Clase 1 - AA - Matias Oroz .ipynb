```python
import numpy as np


print("Ejercicios de Arrays y Matrices con NumPy - Semana 1")

# Ej1: Array de 10 ceros
ej1 = np.zeros(10)
print("\nEj1:", ej1)

# Ej2: Números 1 al 20
ej2 = np.arange(1, 21)
print("\nEj2:", ej2)

# Ej3: Números pares 2 al 50
ej3 = np.arange(2, 51, 2)
print("\nEj3:", ej3)

# Ej5: Matriz 4x4 aleatoria 0-1
ej5 = np.random.rand(4, 4)
print("\nEj5:\n", ej5)

# Ej6: Suma elemento a elemento
a = np.array([1,2,3])
b = np.array([4,5,6])
ej6 = a + b
print("\nEj6:", ej6)

# Ej7: Multiplicación de matrices 3x3
mat7_1 = np.random.randint(1, 11, (3,3))
mat7_2 = np.random.randint(1, 11, (3,3))
ej7 = np.dot(mat7_1, mat7_2)
print("\nEj7 - Mat1:\n", mat7_1)
print("Mat2:\n", mat7_2)
print("Producto:\n", ej7)

# Ej8: Array 15 valores aleatorios 1-100, max/min/promedio
ej8 = np.random.randint(1, 101, 15)
print("\nEj8:", ej8)
print("Max:", np.max(ej8), "Min:", np.min(ej8), "Promedio:", np.mean(ej8))

# Ej9: Matriz 5x5 (1-25), extraer diagonal
ej9 = np.arange(1, 26).reshape(5,5)
diag9 = np.diag(ej9)
print("\nEj9 - Matriz:\n", ej9)
print("Diagonal:", diag9)

# Ej10: Matriz 5x5 aleatoria 1-20, suma filas/columnas
ej10 = np.random.randint(1, 21, (5,5))
print("\nEj10 - Matriz:\n", ej10)
print("Suma filas:", np.sum(ej10, axis=1))
print("Suma columnas:", np.sum(ej10, axis=0))
```
