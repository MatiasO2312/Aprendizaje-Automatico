**Importar librerias y cargar el dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar dataset
df = pd.read_csv("/content/sample_data/Automobile.csv")

# Vista previa
print(df.head())
print(df.columns.tolist())
```

**Gráfico de dispersión: mpg vs weight**

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["weight"], df["mpg"], color="steelblue", alpha=0.6)
plt.title("Relación entre peso y consumo de combustible")
plt.xlabel("Peso del vehículo (weight)")
plt.ylabel("Consumo (mpg)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Histograma del consumo (mpg)**

```python
plt.figure(figsize=(8, 5))
plt.hist(df["mpg"], bins=20, color="mediumseagreen", edgecolor="black")
plt.title("Distribución del consumo de combustible (mpg)")
plt.xlabel("Millas por galón (mpg)")
plt.ylabel("Cantidad de vehículos")
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Boxplot: cilindros por origen**

```python
plt.figure(figsize=(8, 5))

grupos = [df[df["origin"] == o]["cylinders"].dropna() for o in ["usa", "europe", "japan"]]

plt.boxplot(grupos, labels=["USA", "Europa", "Japón"])
plt.title("Cilindros por origen del vehículo")
plt.xlabel("Origen")
plt.ylabel("Cantidad de cilindros")
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Gráfico de barras: autos por año y origen**

```python
cconteo = df.groupby(["model_year", "origin"]).size().unstack(fill_value=0)

cconteo.plot(kind="bar", figsize=(12, 5), color=["steelblue", "tomato", "mediumseagreen"])
plt.title("Distribución de automóviles por año y origen")
plt.xlabel("Año del modelo")
plt.ylabel("Cantidad de autos")
plt.legend(title="Origen")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```
